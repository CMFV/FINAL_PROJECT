```python
def deleting_raw(x):
    if 'raw ' in x:
        x1=x.replace('raw ', '')
        return x1
    elif'.raw' in x:
        x2=x.replace('.raw', '')
        return x2
    elif'. raw' in x:
        x3=x.replace('. raw', '')
        return x3
    else:return x
```

```python
df_properties['Name']=df_properties['Name'].apply(deleting_raw)
```

```python
def to_lowercase(x):
    x=x.lower()
    return x
df_properties['Name']=df_properties['Name'].apply(to_lowercase)
```

```python
def deleting_points(x):
    if '.' in x:
        x1=x.replace('.', ' ')
        return x1
    else: return x
df_properties['Name']=df_properties['Name'].apply(deleting_points)
```

```python
def deleting_bar(x):
    if '-' in x:
        x1=x.replace('-', ' ')
        return x1
    else: return x
df_properties['Name']=df_properties['Name'].apply(deleting_bar)
```

```python
def generateScores(lst, metric, toFind):
    scores = enumerate(map(lambda x :metric(toFind,x), lst)) 
    scores = sorted(scores, key=lambda x:x[1]) 
    index, score = scores[0] 
    return list(map(lambda x: (lst[x[0]], x[1]), scores))[:10]
        
        

metric1 = textdistance.levenshtein.distance 
metric2 = textdistance.hamming.normalized_distance 
metric3 = textdistance.mlipns.normalized_distance 
metric4 = textdistance.damerau_levenshtein.distance 
metric5 = textdistance.jaro_winkler.distance
metric6 = textdistance.cosine.distance

palabras = df_properties['Name'] 
display(generateScores(palabras,metric6,"white sugar"))
```
