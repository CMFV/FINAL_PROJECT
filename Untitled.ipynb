```python
import textdistance

def generateScores(lst, metric, toFind):
    scores = enumerate(map(lambda x :metric(toFind,x), lst))
    scores = sorted(scores, key=lambda x:x[1])
    index, score = scores[0]
    return {
        "most_similar":lst[index],
        "score_similar":score,
        "sorted_list": list(map(lambda x: (lst[x[0]], x[1]), scores))[:5]
    }
palabras = food_components2['alim_nom_eng']

metric1 = textdistance.levenshtein.distance
metric2 = textdistance.hamming.normalized_distance
metric3 = textdistance.mlipns.normalized_distance
metric4 = textdistance.damerau_levenshtein.distance
metric5 = textdistance.jaro_winkler.distance
display(generateScores(palabras,metric1,"raw"))
display(generateScores(palabras,metric2,"raw"))
display(generateScores(palabras,metric3,"raw"))
display(generateScores(palabras,metric4,"raw"))
display(generateScores(palabras,metric5,"raw"))
```
